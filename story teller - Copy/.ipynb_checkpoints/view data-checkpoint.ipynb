```python
import json
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from typing import Dict, List, Any, Union
import warnings
import os
from pathlib import Path
import glob
warnings.filterwarnings('ignore')

class MultiStoryAnalyzer:
    """
    A comprehensive analyzer for comparing keyword extractors across multiple stories
    in story generation with image synthesis evaluation.
    """
    
    def __init__(self, data_sources: Union[str, List[str], List[Dict]]):
        """
        Initialize the analyzer with multiple story data sources.
        
        Args:
            data_sources: Can be:
                - String path to directory containing JSON files
                - List of file paths to JSON files  
                - List of JSON data dictionaries
        """
        self.extractors = ['yake', 'keybert', 'simple']
        self.stories_data = self._load_data(data_sources)
        self.df = self._create_master_dataframe()
        
    def _load_data(self, data_sources: Union[str, List[str], List[Dict]]) -> List[Dict]:
        """Load data from various sources."""
        stories = []
        
        if isinstance(data_sources, str):
            # Directory path - load all JSON files
            directory_path = Path(data_sources)
            if directory_path.is_dir():
                json_files = list(directory_path.glob("*.json"))
                print(f"Found {len(json_files)} JSON files in directory: {directory_path}")
                
                for file_path in json_files:
                    try:
                        with open(file_path, 'r', encoding='utf-8') as f:
                            data = json.load(f)
                            data['source_file'] = file_path.name
                            stories.append(data)
                            print(f"✓ Loaded: {file_path.name}")
                    except Exception as e:
                        print(f"✗ Error loading {file_path}: {e}")
            else:
                raise ValueError(f"Directory not found: {directory_path}")
                
        elif isinstance(data_sources, list):
            if all(isinstance(item, str) for item in data_sources):
                # List of file paths
                for file_path in data_sources:
                    try:
                        with open(file_path, 'r', encoding='utf-8') as f:
                            data = json.load(f)
                            data['source_file'] = Path(file_path).name
                            stories.append(data)
                            print(f"✓ Loaded: {Path(file_path).name}")
                    except Exception as e:
                        print(f"✗ Error loading {file_path}: {e}")
                        
            elif all(isinstance(item, dict) for item in data_sources):
                # List of JSON dictionaries
                for i, data in enumerate(data_sources):
                    data['source_file'] = data.get('source_file', f'story_{i+1}')
                    stories.append(data)
                    print(f"✓ Loaded: {data['source_file']}")
        
        if not stories:
            raise ValueError("No valid story data found!")
            
        print(f"\nTotal stories loaded: {len(stories)}")
        return stories
        
    def _create_master_dataframe(self) -> pd.DataFrame:
        """Create a structured DataFrame from all story JSON data."""
        rows = []
        
        for story_idx, story_data in enumerate(self.stories_data):
            story_id = story_data.get('id', f'story_{story_idx}')
            story_title = story_data.get('title', f'Story {story_idx + 1}')
            source_file = story_data.get('source_file', f'story_{story_idx + 1}.json')
            
            for paragraph in story_data.get('paragraphs', []):
                paragraph_id = paragraph.get('id')
                paragraph_text = paragraph.get('text', '')
                
                for extractor in self.extractors:
                    if extractor in paragraph.get('results', {}):
                        result = paragraph['results'][extractor]
                        
                        # Extract BERT scores safely
                        bert_precision = None
                        bert_recall = None
                        bert_f1 = None
                        
                        if 'bert_score' in result and result['bert_score']:
                            bert_scores = result['bert_score']
                            bert_precision = bert_scores.get('precision', [None])[0] if bert_scores.get('precision') else None
                            bert_recall = bert_scores.get('recall', [None])[0] if bert_scores.get('recall') else None
                            bert_f1 = bert_scores.get('f1', [None])[0] if bert_scores.get('f1') else None
                        
                        # Extract ROUGE scores safely
                        rouge_scores = result.get('rouge_score', {})
                        
                        row = {
                            'story_id': story_id,
                            'story_title': story_title,
                            'source_file': source_file,
                            'paragraph_id': paragraph_id,
                            'extractor': extractor,
                            'clip_score': result.get('clip_score'),
                            'bias': result.get('bias', False),
                            'bert_precision': bert_precision,
                            'bert_recall': bert_recall,
                            'bert_f1': bert_f1,
                            'rouge1_precision': rouge_scores.get('rouge1', {}).get('precision'),
                            'rouge1_recall': rouge_scores.get('rouge1', {}).get('recall'),
                            'rouge1_f1': rouge_scores.get('rouge1', {}).get('f1'),
                            'rouge2_precision': rouge_scores.get('rouge2', {}).get('precision'),
                            'rouge2_recall': rouge_scores.get('rouge2', {}).get('recall'),
                            'rouge2_f1': rouge_scores.get('rouge2', {}).get('f1'),
                            'rougeL_precision': rouge_scores.get('rougeL', {}).get('precision'),
                            'rougeL_recall': rouge_scores.get('rougeL', {}).get('recall'),
                            'rougeL_f1': rouge_scores.get('rougeL', {}).get('f1'),
                            'brisque_score': result.get('brisque_score'),
                            'prompt': result.get('prompt', '')
                        }
                        rows.append(row)
        
        df = pd.DataFrame(rows)
        print(f"\nDataFrame created with {len(df)} rows across {df['story_id'].nunique()} stories")
        print(f"Stories: {list(df['story_title'].unique())}")
        return df
    
    def plot_overall_extractor_comparison(self, figsize=(18, 12)):
        """Create comprehensive comparison plots across all stories."""
        
        plt.style.use('default')
        sns.set_palette("husl")
        
        fig, axes = plt.subplots(2, 3, figsize=figsize)
        fig.suptitle('Keyword Extractor Performance Comparison Across All Stories', 
                    fontsize=16, fontweight='bold')
        
        # 1. CLIP Score comparison
        sns.boxplot(data=self.df, x='extractor', y='clip_score', ax=axes[0,0])
        axes[0,0].set_title('CLIP Score Distribution', fontweight='bold')
        axes[0,0].set_ylabel('CLIP Score')
        
        # 2. BERT F1 Score comparison
        sns.boxplot(data=self.df, x='extractor', y='bert_f1', ax=axes[0,1])
        axes[0,1].set_title('BERT F1 Score Distribution', fontweight='bold')
        axes[0,1].set_ylabel('BERT F1 Score')
        
        # 3. ROUGE-1 F1 Score comparison
        sns.boxplot(data=self.df, x='extractor', y='rouge1_f1', ax=axes[0,2])
        axes[0,2].set_title('ROUGE-1 F1 Score Distribution', fontweight='bold')
        axes[0,2].set_ylabel('ROUGE-1 F1 Score')
        
        # 4. BRISQUE Score comparison
        sns.boxplot(data=self.df, x='extractor', y='brisque_score', ax=axes[1,0])
        axes[1,0].set_title('BRISQUE Score Distribution\n(Lower = Better Image Quality)', fontweight='bold')
        axes[1,0].set_ylabel('BRISQUE Score')
        
        # 5. Bias comparison
        if 'bias' in self.df.columns:
            bias_counts = self.df.groupby(['extractor', 'bias']).size().unstack(fill_value=0)
            if len(bias_counts.columns) > 1:
                bias_percentages = bias_counts.div(bias_counts.sum(axis=1), axis=0) * 100
                bias_percentages.plot(kind='bar', ax=axes[1,1], stacked=True)
                axes[1,1].set_title('Bias Distribution (%)', fontweight='bold')
                axes[1,1].set_ylabel('Percentage')
                axes[1,1].set_xlabel('Extractor')
                axes[1,1].legend(['No Bias', 'Bias'])
                axes[1,1].tick_params(axis='x', rotation=0)
        
        # 6. Performance by Story
        story_means = self.df.groupby(['story_title', 'extractor'])['clip_score'].mean().unstack()
        story_means.plot(kind='bar', ax=axes[1,2])
        axes[1,2].set_title('CLIP Score by Story', fontweight='bold')
        axes[1,2].set_ylabel('Mean CLIP Score')
        axes[1,2].set_xlabel('Story')
        axes[1,2].legend(title='Extractor')
        axes[1,2].tick_params(axis='x', rotation=45)
        
        plt.tight_layout()
        plt.show()
    
    def plot_story_wise_comparison(self, figsize=(20, 12)):
        """Create story-wise comparison plots."""
        
        num_stories = self.df['story_id'].nunique()
        cols = 3
        rows = (num_stories + cols - 1) // cols
        
        fig, axes = plt.subplots(rows, cols, figsize=figsize)
        fig.suptitle('CLIP Score Comparison by Individual Story', fontsize=16, fontweight='bold')
        
        if rows == 1:
            axes = axes.reshape(1, -1)
        
        stories = sorted(self.df['story_title'].unique())
        
        for idx, story in enumerate(stories):
            row, col = idx // cols, idx % cols
            
            story_data = self.df[self.df['story_title'] == story]
            
            sns.boxplot(data=story_data, x='extractor', y='clip_score', ax=axes[row, col])
            axes[row, col].set_title(f'{story}', fontweight='bold', fontsize=10)
            axes[row, col].set_xlabel('Extractor')
            axes[row, col].set_ylabel('CLIP Score')
        
        # Hide empty subplots
        for idx in range(num_stories, rows * cols):
            row, col = idx // cols, idx % cols
            axes[row, col].set_visible(False)
        
        plt.tight_layout()
        plt.show()
    
    def create_comprehensive_performance_table(self) -> pd.DataFrame:
        """Create a comprehensive performance table across all stories."""
        
        stats_list = []
        
        # Overall statistics across all stories
        for extractor in self.extractors:
            extractor_data = self.df[self.df['extractor'] == extractor]
            
            stats = {
                'Extractor': extractor.upper(),
                'Dataset': 'ALL STORIES',
                'Sample Size': len(extractor_data),
                'CLIP Score (μ±σ)': f"{extractor_data['clip_score'].mean():.4f} ± {extractor_data['clip_score'].std():.4f}",
                'BERT F1 (μ±σ)': f"{extractor_data['bert_f1'].mean():.4f} ± {extractor_data['bert_f1'].std():.4f}",
                'ROUGE-1 F1 (μ±σ)': f"{extractor_data['rouge1_f1'].mean():.4f} ± {extractor_data['rouge1_f1'].std():.4f}",
                'ROUGE-2 F1 (μ±σ)': f"{extractor_data['rouge2_f1'].mean():.4f} ± {extractor_data['rouge2_f1'].std():.4f}",
                'BRISQUE (μ±σ)': f"{extractor_data['brisque_score'].mean():.4f} ± {extractor_data['brisque_score'].std():.4f}",
                'Bias Rate (%)': f"{(extractor_data['bias'].sum() / len(extractor_data) * 100):.1f}%"
            }
            stats_list.append(stats)
        
        # Story-wise statistics
        for story in sorted(self.df['story_title'].unique()):
            story_data = self.df[self.df['story_title'] == story]
            
            for extractor in self.extractors:
                extractor_data = story_data[story_data['extractor'] == extractor]
                
                if len(extractor_data) > 0:
                    stats = {
                        'Extractor': extractor.upper(),
                        'Dataset': story,
                        'Sample Size': len(extractor_data),
                        'CLIP Score (μ±σ)': f"{extractor_data['clip_score'].mean():.4f} ± {extractor_data['clip_score'].std():.4f}",
                        'BERT F1 (μ±σ)': f"{extractor_data['bert_f1'].mean():.4f} ± {extractor_data['bert_f1'].std():.4f}",
                        'ROUGE-1 F1 (μ±σ)': f"{extractor_data['rouge1_f1'].mean():.4f} ± {extractor_data['rouge1_f1'].std():.4f}",
                        'ROUGE-2 F1 (μ±σ)': f"{extractor_data['rouge2_f1'].mean():.4f} ± {extractor_data['rouge2_f1'].std():.4f}",
                        'BRISQUE (μ±σ)': f"{extractor_data['brisque_score'].mean():.4f} ± {extractor_data['brisque_score'].std():.4f}",
                        'Bias Rate (%)': f"{(extractor_data['bias'].sum() / len(extractor_data) * 100):.1f}%"
                    }
                    stats_list.append(stats)
        
        return pd.DataFrame(stats_list)
    
    def plot_correlation_heatmap(self, figsize=(14, 10)):
        """Create a correlation heatmap of all numerical metrics."""
        
        numerical_cols = ['clip_score', 'bert_precision', 'bert_recall', 'bert_f1',
                         'rouge1_precision', 'rouge1_recall', 'rouge1_f1',
                         'rouge2_precision', 'rouge2_recall', 'rouge2_f1',
                         'rougeL_precision', 'rougeL_recall', 'rougeL_f1',
                         'brisque_score']
        
        # Filter columns that exist in the dataframe
        existing_cols = [col for col in numerical_cols if col in self.df.columns]
        
        if existing_cols:
            correlation_matrix = self.df[existing_cols].corr()
            
            plt.figure(figsize=figsize)
            mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))
            sns.heatmap(correlation_matrix, mask=mask, annot=True, cmap='RdYlBu_r', center=0,
                       square=True, linewidths=0.5, cbar_kws={"shrink": .8})
            plt.title('Correlation Matrix of Evaluation Metrics (All Stories)', 
                     fontsize=14, fontweight='bold')
            plt.tight_layout()
            plt.show()
    
    def create_extractor_ranking_table(self) -> pd.DataFrame:
        """Create ranking table showing which extractor performs best for each metric."""
        
        metrics = ['clip_score', 'bert_f1', 'rouge1_f1', 'rouge2_f1', 'rougeL_f1', 'brisque_score']
        ranking_data = []
        
        for metric in metrics:
            if metric in self.df.columns:
                # Calculate mean for each extractor
                means = self.df.groupby('extractor')[metric].mean()
                
                # Sort (ascending for brisque_score, descending for others)
                if metric == 'brisque_score':
                    ranking = means.sort_values().index.tolist()  # Lower is better
                else:
                    ranking = means.sort_values(ascending=False).index.tolist()  # Higher is better
                
                ranking_data.append({
                    'Metric': metric.replace('_', ' ').title(),
                    '1st Place': ranking[0].upper(),
                    '2nd Place': ranking[1].upper(),
                    '3rd Place': ranking[2].upper(),
                    'Best Score': f"{means[ranking[0]]:.4f}",
                    'Performance Gap': f"{abs(means[ranking[0]] - means[ranking[1]]):.4f}"
                })
        
        return pd.DataFrame(ranking_data)
    
    def statistical_significance_analysis(self) -> pd.DataFrame:
        """Perform comprehensive statistical analysis."""
        try:
            from scipy import stats
        except ImportError:
            print("Warning: scipy not available for statistical tests")
            return pd.DataFrame()
        
        results = []
        metrics = ['clip_score', 'bert_f1', 'rouge1_f1', 'rouge2_f1', 'brisque_score']
        
        for metric in metrics:
            if metric in self.df.columns:
                # Prepare data for each extractor
                metric_data = {}
                for extractor in self.extractors:
                    data = self.df[self.df['extractor'] == extractor][metric].dropna()
                    if len(data) > 0:
                        metric_data[extractor] = data.values
                
                if len(metric_data) >= 2:
                    # Perform ANOVA
                    f_stat, p_value = stats.f_oneway(*metric_data.values())
                    
                    # Effect size (eta-squared)
                    total_var = np.var(np.concatenate(list(metric_data.values())))
                    between_var = np.var([np.mean(data) for data in metric_data.values()])
                    eta_squared = between_var / total_var if total_var > 0 else 0
                    
                    # Determine best extractor
                    means = {ext: np.mean(data) for ext, data in metric_data.items()}
                    if metric == 'brisque_score':
                        best_extractor = min(means.keys(), key=lambda k: means[k])
                    else:
                        best_extractor = max(means.keys(), key=lambda k: means[k])
                    
                    results.append({
                        'Metric': metric.replace('_', ' ').title(),
                        'F-statistic': f'{f_stat:.4f}',
                        'p-value': f'{p_value:.4f}',
                        'Significant': 'Yes' if p_value < 0.05 else 'No',
                        'Effect Size (η²)': f'{eta_squared:.4f}',
                        'Best Extractor': best_extractor.upper(),
                        'Best Score': f'{means[best_extractor]:.4f}',
                        'Sample Sizes': str({k: len(v) for k, v in metric_data.items()})
                    })
        
        return pd.DataFrame(results)
    
    def export_all_results(self, output_dir: str = "analysis_results"):
        """Export all analysis results to files."""
        
        # Create output directory
        Path(output_dir).mkdir(exist_ok=True)
        
        # Export comprehensive performance table
        perf_table = self.create_comprehensive_performance_table()
        perf_table.to_csv(f"{output_dir}/comprehensive_performance_table.csv", index=False)
        
        # Export ranking table
        ranking_table = self.create_extractor_ranking_table()
        ranking_table.to_csv(f"{output_dir}/extractor_ranking_table.csv", index=False)
        
        # Export statistical analysis
        stats_table = self.statistical_significance_analysis()
        if not stats_table.empty:
            stats_table.to_csv(f"{output_dir}/statistical_analysis.csv", index=False)
        
        # Export raw data
        self.df.to_csv(f"{output_dir}/raw_data_all_stories.csv", index=False)
        
        # Export story summary
        story_summary = self.df.groupby('story_title').agg({
            'paragraph_id': 'count',
            'clip_score': ['mean', 'std'],
            'bert_f1': ['mean', 'std'],
            'brisque_score': ['mean', 'std']
        }).round(4)
        story_summary.to_csv(f"{output_dir}/story_summary.csv")
        
        print(f"\n✓ All results exported to '{output_dir}/' directory")
        print(f"Files created:")
        print(f"  - comprehensive_performance_table.csv")
        print(f"  - extractor_ranking_table.csv")
        print(f"  - statistical_analysis.csv")
        print(f"  - raw_data_all_stories.csv")
        print(f"  - story_summary.csv")

# Main analysis function
def analyze_multiple_stories(data_source: Union[str, List[str], List[Dict]], 
                           output_dir: str = "analysis_results"):
    """
    Complete analysis of multiple story datasets.
    
    Args:
        data_source: Directory path, list of file paths, or list of JSON data
        output_dir: Directory to save results
    """
    
    print("=== MULTI-STORY KEYWORD EXTRACTOR ANALYSIS ===\n")
    
    # Create analyzer
    analyzer = MultiStoryAnalyzer(data_source)
    
    # 1. Show basic statistics
    print("1. DATASET OVERVIEW:")
    print("="*50)
    print(f"Total stories analyzed: {analyzer.df['story_id'].nunique()}")
    print(f"Total paragraphs: {analyzer.df['paragraph_id'].nunique()}")
    print(f"Total samples: {len(analyzer.df)}")
    print(f"Extractors: {', '.join([e.upper() for e in analyzer.extractors])}")
    print("\nStory breakdown:")
    story_counts = analyzer.df.groupby('story_title')['paragraph_id'].nunique()
    for story, count in story_counts.items():
        print(f"  - {story}: {count} paragraphs")
    print()
    
    # 2. Performance tables
    print("2. COMPREHENSIVE PERFORMANCE ANALYSIS:")
    print("="*50)
    perf_table = analyzer.create_comprehensive_performance_table()
    print(perf_table.to_string(index=False))
    print("\n")
    
    # 3. Extractor rankings
    print("3. EXTRACTOR RANKINGS BY METRIC:")
    print("="*50)
    ranking_table = analyzer.create_extractor_ranking_table()
    print(ranking_table.to_string(index=False))
    print("\n")
    
    # 4. Statistical analysis
    print("4. STATISTICAL SIGNIFICANCE ANALYSIS:")
    print("="*50)
    stats_table = analyzer.statistical_significance_analysis()
    if not stats_table.empty:
        print(stats_table.to_string(index=False))
    else:
        print("Statistical analysis unavailable (requires scipy)")
    print("\n")
    
    # 5. Generate visualizations
    print("5. GENERATING VISUALIZATIONS...")
    print("="*50)
    
    # Overall comparison
    analyzer.plot_overall_extractor_comparison()
    
    # Story-wise comparison
    analyzer.plot_story_wise_comparison()
    
    # Correlation analysis
    analyzer.plot_correlation_heatmap()
    
    # 6. Export results
    analyzer.export_all_results(output_dir)
    
    return analyzer

# Example usage
if __name__ == "__main__":

    # Example 2: Load from list of files
    file_paths = [
        
         "experiments/exp_20250901_133923.json",
        "experiments/exp_20250901_151344.json",
         "experiments/exp_20250901_161740.json",
         "experiments/exp_20250901_163647.json",
         "experiments/exp_20250901_164827.json",
        # ... more files
    ]
    analyzer = analyze_multiple_stories(file_paths)
    
    
    print("Multi-story analysis toolkit ready!")
    print("\nUsage examples:")
    print("1. From directory: analyze_multiple_stories('/path/to/json/files/')")
    print("2. From file list: analyze_multiple_stories(['/path/file1.json', '/path/file2.json'])")
    print("3. From data list: analyze_multiple_stories([data1_dict, data2_dict])")
```
